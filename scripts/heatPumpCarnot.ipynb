Let's analyze the coefficient of performance (COP) of a Carnot heat pump. This is the formula:

$$\text{COP} = \frac{1}{1 - \frac{T_\text{L}}{T_\text{H}}},$$
where $T$ is temperature and the H and L subscripts denote the high and low temperature reservoirs, respectively. If we want to heat a home in the winter, it is interesting to calculate the effect of $T_\text{L}$ on COP. We can set $T_\text{H}$ to:

```python
TH_C = 20
TH_K = TH_C + 273.15

TL_C = -200
```

We can then vary $T_\text{C}$. In some parts of the US, it is not uncommon for temperatures to reach 0 $^\circ$F or  - 18 $^\circ$C. Therefore, we can set look at a range of $T_\text{C}$ values from ambient down to - 18 $^\circ$C.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.clf()
TL = np.linspace(TL_C+273.15, TH_K -5, 100) # range of temperatures
TH = np.linspace(TH_K, TH_K, 100) # range of temperatures

COP = 1/(1 - (TL/TH))

#plt.plot(TL, COP, 'r-')
#plt.plot([TL_C+273.15, TH_K], [1, 1], 'k-', lw=1)
plt.loglog(TL, COP, 'r-') 
plt.loglog([TL_C+273.15, TH_K], [1, 1], 'k-', lw=1) # Path 1 to 2



plt.xlabel('$T_L$ (K)')
plt.ylabel('COP (-)')
plt.savefig('cop.png')
```
