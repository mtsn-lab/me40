```python
from pyCalor import thermo as th
```

```python
T1_C = 50
T1_K = T1_C + 273.15

x1 = 1

substance = 'water'
```

```python
st1 = th.state(substance, T=T1_K, x=x1, name="1")
```

```python
st1.plot("pv",isoProp="T") # supported plots are: "pv","Ts","ph"
```
