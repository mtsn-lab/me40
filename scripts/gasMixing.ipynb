# Gas mixtures
$n_i$ is the number of moles of $i$ in the mixture. The total number of moles is:

$$n = \sum _{i=1}^j n_i $$

$m_i$ is the amount of mas of $i$ in the mixture. The total mass is:

$$m = \sum _{i=1}^j m_i $$

A certain gas has an amount of gas in the mixture. We call that the mole ratio:

$$y_i = \frac{n_i}{n} $$

$$1 = \sum_{i=1}^j y_i $$

The mass fraction is then:

$$\text{mf}_i = \frac{m_i}{m} $$

$$1 = \sum_{i=1}^j \text{mf}_i $$

## $p-v-T$ behavior of gas mixtures: Ideal and real gases

For gases at low density, we can use the ideal gas relation. The equation of state for a real gas is:

$$p V = Z R T $$

where $Z$ is the compressibility factor. When ideal gases are mixed, their behavior is usually not affected by the other gases, and a nonreacting mixture of ideal gases behaves usually as an ideal gas. One example is air. 

The prediction of gas mixtures is based on two models: 
* Dalton's law of additive pressures
* Amagat's law of additive volumes

Dalton: The pressure of a gas mixture is equal to the sum of the pressures each gas would exert if it existed alone at the mixture temperature and volume:

```
+--------+   +--------+   +--------+
| Gas A  |   | Gas B  |   | Gas Mix|
| V, T   | + | V, T   | = | V, T   |
| p_A    |   | p_B    |   | p_A+p_B|
+--------+   +--------+   +--------+
```

$$p = \sum_{i=1}^j p_i (T,V) $$

where $p_i$ is the component pressure.

Amagat's law of additive volumes: The volume of a gas mixture is equal to the sum of the volumes each gas would occupy if it existed alone at the mixture temperature and pressure:

```
+--------+   +--------+   +--------+
| Gas A  |   | Gas B  |   | Gas Mix|
| p, T   | + | p, T   | = | p, T   |
| V_A    |   | V_B    |   | V_A+V_B|
+--------+   +--------+   +--------+
```

$$V = \sum_{i=1}^j V_i (T,p) $$

where $V_i$ is the component volume.

## Ideal gas mixtures

We can relate $p_i$ and $V_i$ to $y_i$:

$$\frac{p_i}{p} = \frac{n_i R_u T/V}{n R_u T/V} = \frac{n_i}{n} = y_i $$

$$\frac{V_i}{V} = \frac{n_i R_u T/p}{n R_u T/p} = \frac{n_i}{n} = y_i $$

Therefore:

$$\frac{p_i}{p} = \frac{V_i}{V} = \frac{n_i}{n} = y_i $$

This is valid for ideal-gas mixtures. The quantity $y_i p$ is called the partial pressure and $y_i V$ is called the partial volume. 

## Real gas mixtures
We can treat a gas mixtured as a **pseudo-pure substance** by using an accurate equation of state such as the van der Waals or Redlich-Kwong, and to determine the constant coefficients in terms of the coefficients of the components. For van der Waals and Redlich-Kwong, the two constants for the mixture are determined from:

$$ a = \left(\sum_{i=1}^j y_i a_i^{1/2} \right)^2$$

$$ b = \sum_{i=1}^j y_i b_i $$

Another approach for predicting the behavior of gas mixtures treating the mixture again as a **pseudo-pure substance** is Kay's rule. It was proposed by W. B. Kay in 1936, and involves the use of a pseudo-critical pressure, $p_\text{cr}'$ and temperature $T_\text{cr}'$:

$$p_\text{cr}' = \sum_{i=1}^j y_i p_{cr,i} $$

$$T_\text{cr}' = \sum_{i=1}^j y_i T_{cr,i} $$

An alternative to this is to use the compressibility factor in the ideal gas relation:

$$p V = Z n R_u T $$

The compressibility factor of the mixture is then:

$$ Z = \sum_{i=1}^j y_i Z_i$$

This approach does not account for the influence of the dissimilar molecules on each other though. 



## Properties of gas mixtures: Ideal and Real gases

Recall our definition of the partial molar property:

$$\bar{X}_i = \left(\frac{\partial X}{\partial n_i} \right)_{T,p,n_l \ne n_i} $$

All n are constant except for the $n_i$. 

$$X = \sum_{i=1}^{j} \bar{X}_i n_i $$

Examples:

$$U = \sum_{i=1}^{j} \bar{u}_i n_i $$

$$H = \sum_{i=1}^{j} \bar{h}_i n_i $$

$$S = \sum_{i=1}^{j} \bar{s}_i n_i $$

This may be different in mixtures. Assume: $V = V_1 + V_2$. Why may this not work? Thermal effects. If we mix gases, heat may be released. Think about mixing hydrogen with oxygen with combustion. Chemical bonds (secondary, polar bonds). If they break heat release happens. Think about dissolving an ionic compound like calcium chloride in water. 

For a binary mixture, we can define the total volume using the partial molar volumes, $\bar{V}_A$ and $\bar{V}_B$:

$$ V = n_A \bar{V}_A + n_B \bar{V}_B $$

$$ \frac{V}{n} = y_A \bar{V}_A + y_B \bar{V}_B $$

$$ y_A + y_B = 1 $$

$$ \frac{V}{n} = (1 - y_B) \bar{V}_A + y_B \bar{V}_B $$

$$ \frac{V}{n} = (1 - y_B) \bar{V}_A(T,p,y_B) + y_B \bar{V}_B(T,p,y_B) $$

## Property changes in mixing

$$V_\text{components before mixing} = \sum_{i=1}^j n_i \bar{v}_i $$

$$V_\text{mixture} = \sum_{i=1}^j n_i \bar{V}_i $$

$$\Delta V = V_\text{mixture} - V_\text{components before mixing} = \sum_{i=1}^j n_i \left(\bar{V}_i-\bar{v}_i \right) $$

Consider adding nitrogen to oxygen. This makes the oxygen unable to move into all positions. 

Non-ideal mixing:
$$\bar{V}_i-\bar{v}_i \ne 0$$

$$\Delta U_\text{mixing} = \sum_{i=1}^j n_i \left(\bar{U}_i-\bar{u}_i \right) $$

$$\Delta H_\text{mixing} = \sum_{i=1}^j n_i \left(\bar{H}_i-\bar{h}_i \right) $$

For non-ideal, $\Delta U_\text{mixing}$ and $\Delta H_\text{mixing}$ can be zero or non-zero.

$$\Delta S_\text{mixing} = \sum_{i=1}^j n_i \left(\bar{S}_i-\bar{s}_i \right) $$

$\Delta S_\text{mixing} > 0$ for ideal and non-ideal mixing. Spontaneous demixing is not possible.

### Ideal gas assumption

Under this approximation, the properties of the gas are not influenced by the presence of the other gases and each behaves as if it was alone at the mixture temperature $T$ and $p$. This is known as the Gibbs-Dalton law. Also the enthalpy, internal energy, and specific heats of an ideal gas only depend on temperature and are indpendent of the pressure or the volume of the ideal-gas mixture. The partial pressure of a component in an ideal gas mixture is simply:

$$p_i = y_i p $$

Evaluation of $\Delta h$ and $\Delta u$ is therefore straightforward for an ideal-gas mixture since it only requires knowledge of the initial and ifnal temperatures. Care should be taken for evaluating the change in entropy though since it depends on the pressure or volume as well as the temperature. The entropy cahnge of an individual gas in an ideal-gas mixture can be determined from:

$$\Delta \bar{s}_i = \bar{s}_i(T_2, p_2) - \bar{s}_i(T_1, p_1)  $$

$$\Delta \bar{s}_i = \bar{s}_i^\circ(T_2) - \bar{s}_i^\circ(T_1) - R \ln \frac{p_{i,2}}{p_{i,1}} $$

where $\bar{s}_i^\circ$ is the standard molar entropy. That is, the entropy when it is in the standard state. 

From Dalton's law:

$$p_{i,2} = y_i p_2 $$

$$p_{i,1} = y_i p_1 $$

$$\Delta \bar{s}_i = \bar{s}_i^\circ(T_2) - \bar{s}_i^\circ(T_1) - R \ln \frac{p_{2}}{p_{1}} $$

We can also write the specific heat as:

$$\bar{c}_v = \sum_{i=1}^j y_i \bar{c}_{v,i} $$

We can then write:

$$\delta \bar{u} = \bar{c}_v (T_2 - T_1) $$

We can write the specific heat as:

$$\bar{c}_p = \sum_{i=1}^j y_i \bar{c}_{p,i} $$

We can then write:

$$\delta \bar{h} = \bar{c}_h (T_2 - T_1) $$

The general formula for the change in entropy of an ideal gas assuming constant specific heats is:

$$\Delta \bar{s} = \bar{c}_p \ln \frac{T_2}{T_1} - R \ln \frac{p_2}{p_1} $$

We can write this assuming ideal gas mixing for the $i$-th component as:

$$\Delta \bar{s}_i = \bar{c}_{p,i} \ln \frac{T_2}{T_1} - R \ln \frac{p_2}{p_1} $$

## What are the sources of entropy growth in mixing?

1. Temperature equilibrium ($T_{11} \ne T_{21}, ...$); $T_2$ is uniform. Initial temperature of a separate species is different after mixing. All species have equal temperature $\sigma > 0$
2. Pressure equilibration ($p_{11} \ne p_{21}, ...$); $p_2$ is uniform. Momentum exchange. Friction happens. $\sigma > 0$.
3. For idential temperature and pressure, $T_{i,}$ and $p_{i,}$, mixing itself is irreversible.

$$S = k \ln W $$

where $W$ is probability. $S_\text{max} \sim W_\text{max}$. The entropy growth because the most likely case is total uniform diffusion of products.
