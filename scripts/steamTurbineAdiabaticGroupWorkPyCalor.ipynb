```python
from pyCalor import thermo as th
```

```python
# pressure should be in the units MPa, 
# temperature in the units K, 
# quality a number betwee 0 and 1.
mdot_KgPerS = 5.73
g_MperS2 = 9.81

p1_MPa = 2
w1_MperS = 50
T1_K = 400+273.15
z1_m = 10

p2_MPa = 15/1000
w2_MperS = 180
x2 = 0.9
z2_m = 6

water_1 = th.state('water',p=(p1_MPa,'Mpa'),T=(T1_K,'K'),name='1')
water_2 = th.state('water',p=(p2_MPa,'Mpa'),x=x2,name='2')
```

```python
# s1 = water_1.s; print(s1) # kJ/kg*K
h1 = water_1.h# ; print(h1) # kJ/kg
v1 = water_1.v#; print(v1) # m3/kg
u1 = water_1.u#; print(u1) # kJ/kg

print('h1 = {0:1.2f} kJ/kg'.format(h1))
print('v1 = {0:1.2f} m^3/kg'.format(v1))
print('u1 = {0:1.2f} kJ/kg'.format(u1))
```

```python
# s2 = water_2.s; print(s2) # kJ/kg*K
h2 = water_2.h#; print(h2) # kJ/kg
v2 = water_2.v#; print(v2) # m3/kg
u2 = water_2.u#; print(u2) # kJ/kg

print('h2 = {0:1.2f} kJ/kg'.format(h2))
print('v2 = {0:1.2f} m^3/kg'.format(v2))
print('u2 = {0:1.2f} kJ/kg'.format(u2))
```

```python
h1_JperKg = h1*1000 #convert to SI units
h2_JperKg = h2*1000 #convert to SI units
```

```python
pr = th.process([(water_1,water_2)])
```
