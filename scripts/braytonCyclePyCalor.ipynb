```python
from pyCalor import thermo as th
```

```python
substance = 'air'
```

```python
p1_kPa = 100
T1_C = 30

p2_MPa = 10

T3_C = 1200
p3_MPa = p2_MPa

p4_kPa = p1_kPa
```

```python
st1 = th.state(substance, p=p1_kPa, T=T1_C+273.15, name="1")
```

```python
st2 = th.state(substance, p=(p2_MPa,'Mpa'), s=st1.s, name="2")
```

```python
st3 = th.state(substance, p=(p3_MPa,'Mpa'), T=T3_C+273.15, name="3")
```

```python
st4 = th.state(substance, p=p4_kPa, s=st3.s, name="4")
```

```python
pr = th.process([(st1,st2),(st2,st3),(st3,st4),(st4,st1)])
```

```python
pr.plot('pv')
```
