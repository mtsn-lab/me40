```python
from pyCalor import thermo as th
```

```python
# properties
p2_bar = 4
p4_bar = 3

TH_K = 800
TL_K = 400

substance = 'air'
```

```python
#T1_K = TH_K
#T2_K = T1_K

#T3_K = TL_K
#T4_K = T3_K

#p1_bar = pHigh_bar
#p3_bar = pLow_bar
```

```python
st2 = th.state(substance,T=(TH_K,'K'),p=(p2_bar,'bar'),name='2')
```

```python
st3 = th.state(substance,T=(TL_K,'K'),s=st2.s,name='3')
```

```python
st4 = th.state(substance,T=(TL_K,'K'),p=(p4_bar,'bar'),name='4')
```

```python
st1 = th.state(substance,T=(TH_K,'K'),s=st4.s,name='1')
```

```python
pr = th.process([(st1,st2),(st2,st3),(st3,st4),(st4,st1)])
```

```python
pr.plot('pv')
```
